```python
# importing library 
import matplotlib.pyplot as plt  

# add x values 
x = [3,6,9]  

# add y values 
y= [2,6,10]

# plotting the points  
plt.plot(x, y) 

# Label x 
plt.xlabel('x-axis')

# Label y 
plt.ylabel('y-axis')

# Title 
plt.title('First plot!')

# Show plot 
plt.show() 
```
